```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# find the number of total players
num_players= purchase_data['SN'].nunique()
total_players=pd.DataFrame([num_players], columns=['Total Players'])
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#number of unique items (by Item Name for different games)
game_titles=purchase_data['Item ID'].nunique()
game_titles
```

```python
#average Price 
avg_price= purchase_data['Price'].mean()
avg_price
```

```python
#total Purchases (count of Purchase Id)
total_purch= purchase_data['Purchase ID'].count()
total_purch
```

```python
# Total Revenue (sum of Price col)
total_rev= purchase_data['Price'].sum()
total_rev
```

```python
# Print a Summary table of game_titles, avg_price, total_rev
data= [(game_titles, avg_price, total_purch, total_rev)]
summary_df=pd.DataFrame(data,columns= ['Number of Unique Items', 'Average Price', 'Number of Purchases', 'Total Revenue'])

#format the price and revenu columns to currency
summary_df['Average Price']=summary_df['Average Price'].map('${:,.2f}'.format)
summary_df['Total Revenue']=summary_df['Total Revenue'].map('${:,.2f}'.format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Groupby gender and find the count of each
gender= purchase_data.drop_duplicates(['SN'])
gender= gender.groupby('Gender').count()

#rename, organize
gender= gender.rename(columns= {'Purchase ID' : 'Total Count'})
gender= gender.drop(columns=['SN','Age','Item ID','Item Name','Price'])
gender
```

```python
#calculate and add the percentage to the total number of players
g_total= gender['Total Count'].sum()

#add the percentage column by equation
gender['Percentage of Players']= (gender['Total Count'] / g_total).astype(float).map("{:.2%}".format)
gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchasing Analysis by gender (purchase count, average purchase price, total purchase value)
grouped_purch= purchase_data.groupby('Gender').agg(count=pd.NamedAgg(column='Price', aggfunc='count'),
                                                   mean_price=pd.NamedAgg(column='Price', aggfunc='mean'),
                                                   total_price=pd.NamedAgg(column='Price', aggfunc='sum'))
#rename columns
grouped_purch= grouped_purch.rename(columns={'count': 'Purchase Count', 'mean_price':'Average Purchase Price', 'total_price' : 'Total Purchase Value'})
grouped_purch
```

```python
#add the average price column
grouped_purch['Avg Total Purchase per Person']= (grouped_purch['Total Purchase Value'] / grouped_purch['Purchase Count'])

#format the avg price, tota purchase value,  avg total per person
grouped_purch['Average Purchase Price']=grouped_purch['Average Purchase Price'].map('${:,.2f}'.format)
grouped_purch['Total Purchase Value']=grouped_purch['Total Purchase Value'].map('${:,.2f}'.format)
grouped_purch['Avg Total Purchase per Person']=grouped_purch['Avg Total Purchase per Person'].map('${:,.2f}'.format)
grouped_purch
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#age demographics (create the bins)
purchase_data['Age Ranges']= pd.cut(purchase_data['Age'], bins= [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,99.9], labels= ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])
purchase_data
```

```python
#age range dataframe with percentages
age_df= purchase_data.groupby('Age Ranges')['Age Ranges'].count().reset_index(name='Total Count')
age_df= age_df.set_index('Age Ranges')
age_sum= age_df['Total Count'].sum()
age_df['Percentage of Players']=(age_df['Total Count']/age_sum).astype(float).map("{:.2%}".format)
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create the aggregate chart for the ages based on the binning
grouped_age= purchase_data.groupby('Age Ranges').agg(count=pd.NamedAgg(column='Item ID', aggfunc='count'),
                                                   mean_price=pd.NamedAgg(column='Price', aggfunc='mean'),
                                                   total_price=pd.NamedAgg(column='Price', aggfunc='sum'))
grouped_age
```

```python
#rename the columns of the age aggregate dataframe
purch_analysis= grouped_age.rename(columns={'count' : 'Purchase Count', 'mean_price':'Average Purchase Price', 'total_price' : 'Total Purchase Value'})
purch_analysis
```

```python
#add the avg column
purch_analysis['Avg Total Purchase per Person']=(purch_analysis['Total Purchase Value'] / purch_analysis['Purchase Count'])
purch_analysis
```

```python
#reformat columns
purch_analysis['Avg Total Purchase per Person']= purch_analysis['Avg Total Purchase per Person'].map('${:,.2f}'.format)
purch_analysis['Total Purchase Value']= purch_analysis['Total Purchase Value'].map('${:,.2f}'.format)
purch_analysis['Average Purchase Price']= purch_analysis['Average Purchase Price'].map('${:,.2f}'.format)
purch_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#top spenders aggregate chart (count, avg price, total purchase)
grouped_spenders= purchase_data.groupby('SN').agg(purchase_count=pd.NamedAgg(column='Purchase ID', aggfunc='count'),
                                                   mean_price=pd.NamedAgg(column='Price', aggfunc='mean'),
                                                   total_price=pd.NamedAgg(column='Price', aggfunc='sum'))
grouped_spenders
```

```python
#rename top spenders agg columns
spenders_df= grouped_spenders.rename(columns={'count' : 'Purchase Count', 'mean_price':'Average Purchase Price', 'total_price' : 'Total Purchase Value'})
spenders_df
```

```python
#sort the total purchase column by descending value
spenders_df= spenders_df.sort_values(by='Total Purchase Value', ascending=False)

#reformat and get the top 5
spenders_df['Average Purchase Price']=spenders_df['Average Purchase Price'].map('${:,.2f}'.format)
spenders_df['Total Purchase Value']=spenders_df['Total Purchase Value'].map('${:,.2f}'.format)
spenders_df= spenders_df.nlargest(5, 'purchase_count')
spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#MOST POPULAR ITEMS aggregate chart for purchase count and tota purchase values
grouped_pop= purchase_data.groupby('Item ID').agg(Purchase_Count=pd.NamedAgg(column='Item ID', aggfunc='count'),
                                                   Total_Purchase_Value=pd.NamedAgg(column='Price', aggfunc='sum'))
grouped_pop
```

```python
#take the item names and individal prices from the original purchase data, drop uneeded col
purch_pop=purchase_data.drop(['Purchase ID','SN','Age','Gender', 'Age Ranges'], axis = 1)
purch_pop
```

```python
#merge the grouped_pop and purch_pop
pop_merge = grouped_pop.merge(purch_pop, on="Item ID")
pop_merge
```

```python
#reset index as item id of the merged df and drop all duplicate rows
pop_merge= pop_merge.set_index('Item ID')
pop_merge= pop_merge.drop_duplicates()

#reformat
pop_merge['Price']=pop_merge['Price'].map('${:,.2f}'.format)
pop_merge= pop_merge.reindex(columns=['Item Name', 'Purchase_Count', 'Price', 'Total_Purchase_Value'])
pop_merge
```

```python
#order by deceasing value for purchase count col
popular= pop_merge.sort_values(by='Purchase_Count', ascending=False)
popular
```
