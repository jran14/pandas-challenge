```python
 # Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

student_data.head()
```

```python
school_data .head()
```

```python
#total school count
total_schools=school_data['School ID'].count()
total_schools
```

```python
#total students
total_students=student_data['Student ID'].count()
total_students
```

```python
#total budget 
total_budget=school_data['budget'].sum()
total_budget
```

```python
#avg math (mean)
number_of_students=student_data['math_score'].count()
sum_math= student_data['math_score'].sum()
average_math= sum_math / number_of_students
average_math
```

```python
#avg reading (mean)
sum_read= student_data['reading_score'].sum()
average_read= (sum_read / number_of_students) 
average_read
```

```python
#sum of >70 math scores (mean)
pass_math=(student_data['math_score']>= 70).mean()
pass_math= pass_math*100
pass_math
```

```python
#sum of >70 reading scores (mean)
pass_read=(student_data['reading_score']>= 70).mean()
pass_read= pass_read*100
pass_read
```

```python
#avg of > passing math and passing reading scores
overall_pass=(pass_math +pass_read ) /2
overall_pass= overall_pass
```

```python
#creating the district summary dataframe
data= [(total_schools, total_students, total_budget, average_math, average_read, pass_math,pass_read,overall_pass)]
district_summary=pd.DataFrame(data,columns= ['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate'])

#format 
district_summary['Total Budget']=district_summary['Total Budget'].map('${:,.2f}'.format)
district_summary
```
